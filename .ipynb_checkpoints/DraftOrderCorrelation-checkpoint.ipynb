```python
# Dependencies
import requests
import pandas as pd
```

```python
# Define League ID and year
league_id = 730129
year = 2021
```

```python
# Define the URL with our parameters
url = f'https://fantasy.espn.com/apis/v3/games/ffl/leagueHistory/{league_id}/?seasonId={year}'

# Pull team data from the URL 
performance_response = requests.get(url, params={"view": "mTeam"})

# Transform the response into a JSON file
performance_json = performance_response.json()[0]
```

```python
# Initialize empty lists to hold ids, W %
ids = []

win_percentage = []
```

```python
# Loop through the JSON and populate ids and W % lists
for i in range(0,10):
    ids.append(performance_json['teams'][i]['id'])
    win_percentage.append(performance_json['teams'][i]['record']['overall']['percentage'])
```

```python
# Create dictionary to hold performance data
data = {'Team ID' : ids, 'Win Percentage' : win_percentage}

# Load the dictionary into the DataFrame
performance_df = pd.DataFrame(data)
```

```python
performance_df
```

```python
# Reuse the URL variable to call the draft view
draft_response = requests.get(url, params={"view" : "mDraftDetail"})

# Convert the response to a JSON
draft_json = draft_response.json()[0]
```

```python
# Initialize the empty lists for ids and draft position
ids = []

draft_position = []
```

```python
# Loop throught JSON and populate the ids and draft position lists
for i in range(0,10):
    ids.append(draft_json['draftDetail']['picks'][i]['teamId'])
    draft_position.append(draft_json['draftDetail']['picks'][i]['id'])
```

```python
# Create a dictionary to hold the draft data
data = {'Team ID' : ids, 'Draft Position' : draft_position}

# Load the dictionary into a pandas DataFrame
draft_df = pd.DataFrame(data)
```

```python
draft_df
```

```python
df = pd.merge(draft_df, performance_df, how='left')
```
